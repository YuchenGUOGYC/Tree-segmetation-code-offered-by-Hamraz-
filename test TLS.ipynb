```python
'''
number_of_trees = 559
for i in range(number_of_trees):
'''   
i = 13
m = 'TLS/TLS_POINTcloud_000' + '{0:03}'.format(i) + '.txt'
file_tree = np.loadtxt(m)
Z = []
X = []
Y = []
for i in range(len(file_tree)):
    z = file_tree[i][2]
    Z.append(z)
    x = file_tree[i][0]
    X.append(x)
    y = file_tree[i][1]
    Y.append(y)
```

```python
height = np.max(Z)-np.min(Z)
gap = 0.1
gap_number =math.ceil(height /gap)
Zmin =np.min(Z)
n = 0

value_group = []
width_group_X = []
width_group_Y = []
while n < gap_number:
    i_group = []
    N = Zmin + gap*n   
    M = Zmin + gap*(n+1)
    for i in range(len(Z)):
        if  Z[i] < M and Z[i] >= N: 
            i_group.append(i) #choose all the Z_values' numbers in area (N,M) 
            value_group.append(Z[i])
    
    X_new = []
    Y_new = []
    for j in range(len(i_group)):
        X_new.append(X[i_group[j]])
        Y_new.append(Y[i_group[j]])
    X_max = np.max(X_new)
    X_min = np.min(X_new)
    Y_max = np.max(Y_new)
    Y_min = np.min(Y_new)
    width_X = X_max - X_min
    width_Y = Y_max - Y_min
    width_group_X.append(width_X)
    width_group_Y.append(width_Y)
    n = n + 1   
```

# Find the ground point of TLS

```python
number_of_trees = 559
Z_ground = []
X_ground = []
Y_ground = []
for i in range(number_of_trees):
  
    m = 'TLS/TLS_POINTcloud_000' + '{0:03}'.format(i) + '.txt'
    file_tree = np.loadtxt(m)
    
    
    Z = []
    ground_individual_tree_number = [] 
    for i in range(len(file_tree)):
        z = file_tree[i][2]
        Z.append(z)
    ground_base = np.min(Z)
    for j in range(len(Z)):
        if Z[j] < (ground_base + 0.01):
            ground_individual_tree_number.append(j)
            Z_ground.append(Z[j]) 
    for k in range(len(ground_individual_tree_number)):
        X_ground.append(file_tree[ground_individual_tree_number[k]][0])
        Y_ground.append(file_tree[ground_individual_tree_number[k]][1])      
```

```python
import os, sys
from scipy import stats
import numpy as np
import pandas as pd


x_ground = np.array(X_ground) 
y_ground = np.array(Y_ground)
z_ground = np.array(Z_ground)
df = pd.DataFrame({"X" : x_ground, "Y" : y_ground, "Z": z_ground})
df.to_csv("TLS_ground.csv", index=False)
```
